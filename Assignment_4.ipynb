```python
from sklearn.preprocessing import LabelEncoder
```

```python
LabelEncoder = LabelEncoder()
df['ExtracurricularActivities'] = LabelEncoder.fit_transform(df['ExtracurricularActivities'])
df
```

```python
categorical_cols = df.select_dtypes(include= 'O').columns.tolist()
print(categorical_cols)
```

```python
from sklearn.preprocessing import LabelEncoder

LE = LabelEncoder()
for col in categorical_cols:
  df[col]= LE.fit_transform(df[col])
```

```python
df.head()
```

```python
numerical_cols = df.select_dtypes(exclude= 'O').columns.tolist()
print(numerical_cols)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df[numerical_cols] = scaler.fit_transform(df[numerical_cols])
df.head()
```

```python
df[numerical_cols] = df[numerical_cols].round(3)
df.head()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr().round(2), annot=True)
```

```python
df.columns
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))
sns.scatterplot(x=df['AptitudeTestScore'], y=df['PlacementStatus'])
plt.show()
```

```python
x = df.drop(columns=['StudentID', 'PlacementStatus'])
y = df['PlacementStatus']
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size= 0.20, random_state= 0)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

y_train = pd.cut(y_train, bins=2, labels=[0,1])
y_test = pd.cut(y_test, bins=2, labels=[0,1])

model = LogisticRegression(max_iter=1000)
model.fit(x_train, y_train)

y_pred = model.predict(x_test)

accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy_Score: {accuracy*100: .2f}%")
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree

clf = DecisionTreeClassifier(max_depth= 3)
clf.fit(x_train, y_train)

y_pred = clf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score

DT_accuracy = accuracy_score(y_test, y_pred)
print(f"Decision_tree_accuracy: {DT_accuracy*100: .2f}%")
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15,10))
plot_tree(clf, filled= True, feature_names= x_train.columns, rounded= True)
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier

clf = KNeighborsClassifier()
clf.fit(x_train, y_train)

clf.score(x_test, y_test)

print(f"KNN_accuracy: {clf.score(x_test, y_test)*100: .2f}%")
```

```python
accuracy = []

for k in range(1,11):
  knn = KNeighborsClassifier(n_neighbors= k)
  knn.fit(x_train, y_train)
  acc = knn.score(x_test, y_test)
  accuracy.append(acc)
```

```python
for k, acc in enumerate(accuracy, start=1):
  print(f"K: {k}, Accuracy: {acc*100: .2f}%")
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,11), accuracy, marker='o')
plt.xlabel('Number of Neighbors (K)')
plt.ylabel('Accuracy')
plt.title('Accuracy vs. Number of Neighbors (K)')
plt.show()
```
